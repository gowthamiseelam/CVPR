```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('flower.webp') 
inverse = 255 - image
plt.imshow(inverse)
```

```python
image = cv2.imread('pink.jpeg') 
inverse = cv2.bitwise_not(image)
plt.imshow(inverse)
```

Log Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('pink.jpeg')
c = 255 / np.log(1 + np.max(image))  
log_image = c * (np.log(image + 1))
log_image = np.array(log_image, dtype=np.uint8)
plt.imshow(log_image)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('pink.jpeg')
c = 56
log_image = c * (np.log(image + 1))
log_image = np.array(log_image, dtype=np.uint8)
plt.imshow(log_image)
```

PowerLa(Gamma)

```python
img = cv2.imread('pink.jpeg') 
gamma1, gamma2 = 1.0, 2.5 
gamma_corrected1 = np.array(255*(img / 255) ** gamma1, dtype = 'uint8') 
plt.imshow(gamma_corrected1) 
gamma_corrected2 = np.array(255*(img / 255) ** gamma2, dtype = 'uint8') 
plt.imshow(gamma_corrected2)
```

PieceWise-Linear 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('pink.jpeg') 
def piecewise_linear(image, t1, t2, s1, s2):
    output = np.zeros_like(image)
    output[image <= t1] = (image[image <= t1] / t1) * s1
    output[(image > t1) & (image <= t2)] = s1 + ((image[(image > t1) & (image <= t2)] - t1) / (t2 - t1)) * (s2 - s1)
    output[image > t2] = s2 + ((image[image > t2] - t2) / (255 - t2)) * (255 - s2)
    
    return np.clip(output, 0, 255)
t1, t2 = 85, 170  
s1, s2 = 50, 200
transformed_image = piecewise_linear(image, t1, t2, s1, s2)
plt.imshow(transformed_image)
```
